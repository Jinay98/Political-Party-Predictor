# importing the libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn import tree
%matplotlib inline

# do not display warnings in notebook 
import warnings
warnings.filterwarnings('ignore')
```

# Reading the data from the csv file

```python
train=pd.read_csv('train.csv')
train.head()
train.shape
```

# Visualizing the data

```python
train.groupby(['YOB','Party'])['Party'].count()

train.groupby(['Income','Party'])['Party'].count()
```

# Dropping the columns 'Unnamed' and User id
because those have no relevance to the party 

```python
train.drop(train.columns[[0,1]], axis=1, inplace=True)
train.shape
```

# Checking for the null values in each column

```python
train.isnull().sum()
```

# Year of Birth is replaced by age
The age is a better factor to evaluate as compared to the year

```python
import datetime
current_year = datetime.datetime.now().year
print(current_year)
train["YOB"] = train["YOB"].apply(lambda x: current_year - x)
train['YOB']
train.rename(columns = {"YOB": "Age"},inplace=True) 
train.head()
```

# Label Encoding 
Male is replaced by 1
Female is replaced by 0
Yes is replaced by 1
No is replaced by 0
under $25,000:1
$25,001 - $50,000:2
$50,000 - $74,999:3
$75,000 - $100,000:4
$100,001 - $150,000:5
over $150,000':6

```python
train.replace(to_replace ="Male", 
                 value =0,inplace=True)
train.replace(to_replace ="Female", 
                 value =1,inplace=True)
train.replace(to_replace ="Yes", 
                 value =1,inplace=True)
train.replace(to_replace ="No", 
                 value =0,inplace=True)
train['Income'] = train['Income'].map( {'under $25,000':1, 'over $150,000':6,  '$75,000 - $100,000':4,'$50,000 - $74,999':3,'$25,001 - $50,000':2, '$100,001 - $150,000':5 } )
train.head()
```

# Dropping the rows that dont have atleast 10 not null values

```python
train.dropna(thresh=10,inplace=True)
train.shape
```

# Replacing the NaN values with the mode excluding the NaN
    hence all the Nan values are replaced by the most frequent feature of that column

```python
col=train.columns
#col=['Income','Gender','Q124742','Q124122','Q123464','Q123621','Q122769','Q122770','Q122120','Q121699','Q121700','Q120978',
#    'Q121011','Q120379','Q120650','Q120012','Q120014','Q119334','Q119851','Q118892','Q118117',
#    'Q118233','Q118237']
for i in col:
    train[i].fillna(train[i].value_counts().idxmax(),inplace=True)
train.head()
train.isnull().sum()
```

# Label encoding the householdStatus and the Education level
While mapping education level we set weights as per higher degree of education

for Q122771 the question onlypertains to K-12 education so we only allocate public and private values to them otherwise we put 0

```python
set(train['Party'])
train['HouseholdStatus'] = train['HouseholdStatus'].map( {'Domestic Partners (no kids)':1, 'Domestic Partners (w/kids)':2,  'Married (no kids)':3,'Married (w/kids)':4,'Single (no kids)':5,'Single (w/kids)':6 })
train.head(25)
#train.isnull().sum()
set(train['EducationLevel'])
train['EducationLevel'] = train['EducationLevel'].map( {'Current K-12':1,'High School Diploma':2, 'Current Undergraduate':3,  'Associate\'s Degree':4,'Bachelor\'s Degree':5,'Master\'s Degree':6,'Doctoral Degree':7 })
train['EducationLevel']
train.head()
col=['EducationLevel']
train['EducationLevel']
train.head()
train.shape
# for index, row in train.iterrows():
#     if row['EducationLevel']==1:
#         train['Q122771']=train['Q122771'].map( {'Public':1,'Private':2 })
#     else:
#         train['Q122771']=train['Q122771'].map( {'Public':0,'Private':0 })
        
# # for i in col:
# #     if train['EducationLevel']==1:
# #         train['Q122771']=train['Q122771'].map( {'Public':1,'Private':2 })
# #     else:
# #         train['Q122771']=train['Q122771'].map( {'Public':0,'Private':0 })
# train['Q122771']        
```

# Label Encoding 
hotheaded isbad then cool heaaded same with standard and odd hours also giving is much better than receiving thus these all parameters have higher weights

We identify several questions that are totally unrelated to the party of the candidate
therefore to eliminate those redundancies we remove the columns

Also we use OneHotEncoding for columns where we dont want a particular bias to an answer in those cases we use one hot encoding

```python
set(train['Q117186'])
train['Q117186'] = train['Q117186'].map( {'Cool headed':2,'Hot headed':0})
set(train['Q117193'])
train['Q117193'] = train['Q117193'].map( {'Standard hours':2,'Odd hours':1})
train['Q117193']
set(train['Q119650'])
train['Q119650'] = train['Q119650'].map( {'Giving':1,'Receiving':-1})
train['Q98059'] = train['Q98059'].map( {'Yes':1,'Only-child':0})
train['Q102089'] = train['Q102089'].map( {'Own':1,'Rent':0})
train['Q106997'] = train['Q106997'].map( {'Grrr people':-1,'Yay people!':1})
train['Q108950'] = train['Q108950'].map( {'Cautious':1,'Risk-friendly':-1})
train['Q116197'] = train['Q116197'].map( {'P.M.':1,'A.M.':0})
train['Q122771']=train['Q122771'].map( {'Public':1,'Private':2 })
train['Q120472']=train['Q120472'].map( {'Art':1,'Science':0 })
train['Q101162']=train['Q101162'].map( {'Pessimist':-1,'Optimist':1 })

train.head()
train.drop(['Q98059','Q120194','Q115899','Q99982','Q101163','Q108855','Q108342','Q108855','Q108856','Q110740','Q114386','Q113583','Q113584','Q111580',],axis=1,inplace=True)
train.shape

train = pd.get_dummies(train, columns=['Q118232','Q116881','Q115777'])
g = train.columns.to_series().groupby(train.dtypes).groups

print(g)
# train['Q118232']
# print()
# print(train.dtypes)
```

```python
train['Party']=train['Party'].map( {'Republican':1,'Democrat':0 })
```

```python
list(train.dtypes)
train.head(100)
```

# Moving parties to the right most


```python
cols_at_end = ['Party']
train = train[[c for c in train if c not in cols_at_end] 
        + [c for c in cols_at_end if c in train]]
train.head()
# train.shape
```

```python
y_train = train['Party']
x_train = train.drop(['Party'], axis=1).values
```

```python
# Create validation dataset out of training data
val_share = int(0.15*len(y_train))

x_val = x_train[-val_share:]
y_val = y_train[-val_share:]

x_train = x_train[:-val_share]
y_train = y_train[:-val_share]
```

```python
train.nunique()
```

```python
train.head()
```

# Using Decesion Tree for Classification
15 % of the training data is used as validation model and then the accuracy is calculated

```python
decision_tree = tree.DecisionTreeClassifier(max_depth=5)
decision_tree.fit(x_train, y_train)
y_pred = decision_tree.predict(x_train)
acc_decision_tree_train = decision_tree.score(x_train, y_train) * 100
print("Training Accuracy: %0.3f" % acc_decision_tree_train)

acc_decision_tree_val = decision_tree.score(x_val, y_val) * 100
print("Validation Accuracy: %0.3f" % acc_decision_tree_val)
# print(y_pred)
# print(y_train)
```

# Representing the Accuracy of the Model
through confusion matrix,ROC curve

```python
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score
import seaborn as sns
cm = confusion_matrix(y_pred,y_train)
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cm), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
tn, fp, fn, tp = cm.ravel()
```

```python
precision = tp/(tp+fp) 
recall = tp/(tp+fn) 
print(precision, recall)
```

# Using Random forest Classification to improve the accuracy
We find out the best features and using those features we classify

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import confusion_matrix, recall_score, precision_score
```

```python
n_estimators = np.arange(3,25,2)
max_depth = np.arange(1,5,1)
criterion = ['gini', 'entropy']
min_samples_split = np.arange(2,6,1)
min_samples_leaf = np.arange(2,6,1)
random_state = [1337]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth, 
               'criterion': criterion,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'random_state': random_state}

print(random_grid)
```

```python
rf = RandomForestClassifier()

# Instantiate Random Search of Parameters using 3 fold cross validation
rf_random = RandomizedSearchCV(estimator=rf,
                               param_distributions=random_grid, 
                               n_iter=10, 
                               cv=4, 
                               random_state=137)
```

```python
rf_random.fit(x_train, y_train) 
print('Done')
```

```python
rf_random.best_params_
```

```python
best_rf = rf_random.best_estimator_
predictions = list(best_rf.predict(x_train))

print('Accuracy:', best_rf.score(x_train, y_train))
print('Precision:', precision_score(list(y_train), predictions))
print('Recall:', recall_score(list(y_train), predictions))
```
